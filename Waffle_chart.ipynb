**Question:** Create a Waffle chart to dispaly the proportiona of China and Inida total immigrant contribution.


```python
df_ci = df_can.loc[['India', 'China'], :]

fig = plt.figure(FigureClass=Waffle, rows=20, columns=20,
                 values=df_ci['Total'],
                 cmap_name='Set2',
                 legend={
                         "loc": 'upper left',
                         'bbox_to_anchor': (0, -0.1),
                         "ncol": 2})

legend_labels = [f"{country} ({value})" for country, value in zip(df_ci.index, df_ci['Total'])]
                                                                    # india   , china
legend_handles = [mpatches.Patch(color=c, label=l) for c, l in zip(['#33a02c', '#ff7f00'], legend_labels)] # Specify colors here
plt.legend(handles=legend_handles, loc='upper left', bbox_to_anchor=(1, 1))
plt.show()
```
